```python
df["SpecFile"] = df["reference_file_name"] + ".mzML"
df["opt_global_cv_MS:1000889_peptidoform_sequence"] = df["peptidoform"].str.replace("[","(")
df["opt_global_cv_MS:1000889_peptidoform_sequence"] = df["opt_global_cv_MS:1000889_peptidoform_sequence"].str.replace("]",")")
df["opt_global_cv_MS:1000889_peptidoform_sequence"] = df["opt_global_cv_MS:1000889_peptidoform_sequence"].str.replace("-","")
df["ScanNum"] = df["scan_number"]
```

```python
def get_position(p):
    pattern = r'\d+'
    if p == "non-canonical":
        return "non-canonical"
    elif p == "canonical":
        return "canonical"
    else:
        match = re.search(pattern, p)
        return match[0]
    
def get_type(p):
    if p == "non-canonical":
        return "novel"
    elif p == "canonical":
        return "canonical"
    else:
        return p.replace("'","").split("-")[0]
```

```python
df["Variation type"] = df["position"].apply(get_type)
df["position"] = df["position"].apply(get_position)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob
```

### tissue Count

```python
df = pd.read_table("/mnt/nfs/wangd/project/paper_db/count/tissue_peps.tsv")
index = df.columns[0]

df_sorted = df.sort_values('Count', ascending=False)
plt.bar(df_sorted[index], df_sorted['Count'])
plt.xlabel(index)
plt.ylabel('Count')
plt.title('Peptide_Count')

plt.xticks(rotation='vertical')  
plt.tight_layout()  

for i, count in enumerate(df_sorted['Count']):
    plt.text(i, count, str(count), ha='center', va='bottom', fontsize=7)

plt.savefig('/mnt/nfs/wangd/project/paper_db/count/count-plot/' + index + '-plot.svg', format='svg')
```
